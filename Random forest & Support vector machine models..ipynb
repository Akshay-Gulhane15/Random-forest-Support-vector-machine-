```python
#a)	What is the accuracy if the problem is solved using Random Forest model?

# STEP 1 - Importing Libraries and our Data
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
air=pd.read_excel((r'C:\Users\ASUS\Desktop\data sets\Airline Satisfaction.xlsx'))
```

```python
air.head()
```

```python
# Check the Data type and missing values

air.info()
air.isnull().sum()
```

```python
# Summary Statistics of DataSet

air.describe()
```

```python
#changeing (SATISFACTION) from itnger to categorical data

air['satisfaction']=air['satisfaction'].astype('category')
```

```python
air.info()
```

```python
#Arrival Delay in Minutes is a Numerical column which has 26 missing values.

air['Arrival Delay in Minutes'].skew()
```

```python
# replace all the missing values with the Median value (because skewness is > 1 )

air['Arrival Delay in Minutes'].fillna(air['Arrival Delay in Minutes'].median(),inplace=True)
```

```python
air.isnull().sum()
```

```python
new_col = ['Gender','Customer Type','Type of Travel','Class']
```

```python
#creating dummy variables

air_1 = pd.get_dummies(air, columns=new_col, drop_first=True)
```

```python
air_1.head()
```

```python
# STEP 2- Identify X and Y

Y = air_1[['satisfaction']]

X = air_1.drop(columns=['id','satisfaction']) 

# STEP 3- Split the data into training data(~80%) and test data(20%)

from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, train_size=0.8, random_state=4321)

len(X_train), len(X_test), len(Y_train), len(Y_test)
```

```python
#STEP 4- Building the model on Training data

from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=500)

## Fit the object on training data

model = rf.fit(X_train, Y_train)
model
print('The model has run successfuly')
```

```python
# STEP 5-Predicting the values based on test data

Y_test['predicted_satisfaction'] = model.predict(X_test)
```

```python
Y_test.head()
```

```python
# STEP 6- Calculating accuracy by Building Confusion_Matrix, Accuracy score, Classification_report

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

accuracy_RF =accuracy_score(Y_test['satisfaction'],Y_test['predicted_satisfaction'])
```

```python
confusion_matrix(Y_test['satisfaction'],Y_test['predicted_satisfaction'])
```

```python
print(classification_report(Y_test['satisfaction'],Y_test['predicted_satisfaction']))
```

Our Random forest classifier model has 94% accuracy 

```python
# Identifyng the importance of input variable: which input variable is most useful in predicting Y
model.feature_importances_
```

```python
X_train.columns
```

```python
# arrange the importances as per the significance (in descending order)

pd.Series(model.feature_importances_,X_train.columns).sort_values(ascending=False)
```

From above Food and drink, Departure Delay in Minutes, Gender_Male, Class_Eco Plus are least significant input variables from our dataset.
